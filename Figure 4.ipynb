```python

#1 DEFINE CONNECTIVITY AND INPUTS

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import special
from scipy.stats import norm
from scipy.stats import pearsonr
from scipy import signal
from scipy.fft import fftshift

import networkx as nx
import numpy as np
import pandas as pd

T = 50000
dt=0.1
N= 100
d=-1
beta=50
D =0.000
rho =0.05 #connectoin probability
f=0.8#proportion of e vs i cells/cortical
sigma_e_2 = 0.005
sigma_i_2 = 0.005
mu_e= 0.08 #0.08 so that the variance scales like 1/N
mu_i= -f/(1-f)*mu_e
sigma_w2 =  rho*(f*sigma_e_2+(1-f)*sigma_i_2+f/(1-f)*mu_e**2)
B=-.05



#arrays for network dynamics 
time=np.arange(0,T,1)
#environmental perturbations




sigma_h2 =np.logspace(-6,2,9) #np.ones(10)*0.001
Trials =10
var_perturbation = 0.2# np.arange(0,0.4,0.01)
Gamma = np.zeros(T)
Mean_transitions = np.zeros(len(sigma_h2))
Var_transisitons = np.zeros(len(sigma_h2))
number_of_transitions=np.zeros((len(sigma_h2),Trials))
for s in range(len(sigma_h2)):
  for q in range(Trials):
    gamma = 1+4*beta**2*sigma_h2[s]**2
    perturbations = np.zeros(T)
    filtered_noise_1 = np.zeros(T)
    filtered_noise_2 = np.zeros(T)
    filtered_noise_3 = np.zeros(T)
    for t in range(len(time)-1):
      perturbations[t+1] = perturbations[t]+dt*(-perturbations[t]+filtered_noise_1[t])
      filtered_noise_1[t+1]= filtered_noise_1[t]+dt*0.1*(-1*filtered_noise_1[t]+filtered_noise_2[t])
      filtered_noise_2[t+1]= filtered_noise_2[t]+dt*0.1*(-1*filtered_noise_2[t]+filtered_noise_3[t])
      filtered_noise_3[t+1]= filtered_noise_3[t]+dt*(-1*filtered_noise_3[t])+np.sqrt(2*var_perturbation*dt)*np.random.normal(0,1)
    number_of_transitions[s][q]=0
    for t in range(len(time)):
      Gamma[t] =  np.sqrt(N*sigma_w2*beta**2/(np.pi*np.sqrt(gamma))*np.exp(-beta**2*(B+ perturbations[t])**2/(gamma))**2)
      if ((Gamma[t-1]<1 and Gamma[t]>1) or (Gamma[t-1]>1 and Gamma[t]<1)):
        number_of_transitions[s][q]=number_of_transitions[s][q]+1/(T*dt)
      
  Mean_transitions[s] = np.mean(number_of_transitions[s][:])
  Var_transisitons[s] = np.var(number_of_transitions[s][:])

# plt.plot(time, Gamma)
# plt.show()


plt.figure(figsize=(10,5))
plt.rcParams.update({'font.size': 22})
for pos in ['right', 'top', 'bottom', 'left']:
   plt.gca().spines[pos].set_visible(False)
plt.gca().set_prop_cycle(plt.cycler('color', plt.cm.Blues(np.linspace(0, 1, 3))))
for pos in ['right', 'top', 'bottom', 'left']:
   plt.gca().spines[pos].set_visible(False)

plt.scatter(sigma_h2,Mean_transitions, linewidth=4, c=Mean_transitions, cmap='Blues')
plt.fill_between(sigma_h2, Mean_transitions - np.sqrt(Var_transisitons), Mean_transitions + np.sqrt(Var_transisitons),
                 color='gray', alpha=0.2)
plt.xscale("log")
plt.xlim(1e-6,110)
```
